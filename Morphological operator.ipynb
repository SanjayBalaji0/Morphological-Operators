```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img =np.zeros((600,800))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='Sanjay',org=(50,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(5,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    plt.show()
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
kernel=np.ones((5,5),dtype=np.uint8)
```

```python
erosion1=cv2.erode(img,kernel,iterations=4)
```

```python
display_img(erosion1)
```

```python
dilu=cv2.dilate(img,kernel,iterations=4)
display_img(dilu)
```

```python
img=load_img()
white_noise=np.random.randint(low=0,high=2,size=(600,800))
noise_img=img+white_noise
display_img(noise_img)
```

```python
opening=cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
display_img(opening)
```

```python
img=load_img()
black_noise=np.random.randint(low=0,high=2,size=(600,800))
black_noise=black_noise*-255
black_noise_img=img+black_noise
black_noise_img[black_noise_img==-255]=0
display_img(black_noise_img)
```

```python
img = load_img()
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
display_img(gradient)
```
